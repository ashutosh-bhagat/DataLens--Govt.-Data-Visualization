```python
## Here we have created a pivot view
## The pivot function in Pandas is used to reshape data, making it easier for analysis and visualization. It transforms a long-format dataframe into a wide-format table, where one column's values become new columns.
pivot_df = df.pivot_table(index="State_Name", columns="Crop", values="Production", aggfunc="sum", fill_value=0)

## top_state ma largest 10 states store karavie chee
top_states = pivot_df.sum(axis=1).nlargest(10).index.tolist()

top_crops = pivot_df.sum(axis=0).nlargest(10).index.tolist()

pivot_df = np.log1p(pivot_df)
pivot_df = pivot_df.loc[top_states, top_crops]

# Create checkbox widgets for crops
crop_selector = pn.widgets.CheckBoxGroup(name="Select Crops", options=top_crops, value=top_crops, align=('start','start'))
state_selector = pn.widgets.CheckBoxGroup(name="Select State", options=top_states, value=top_states, align=('end','end'))

# Function to plot the stacked bar chart
def plot_stacked_bar(selected_crops, selected_states):
    if not selected_crops:
        return "Please select at least one Crop."

    if not selected_states:
        return "Please select at least one State."
    # Filter data based on selected crops
    filtered_df = pivot_df.loc[selected_states, selected_crops]

    # Create figure
    fig, ax = plt.subplots(figsize=(10, 6))
    
    # Plot stacked bar chart
    filtered_df.plot(kind="bar", stacked=True, ax=ax, colormap="viridis", align='center')

    ax.set_title("Top 10 States by Crop Production")
    ax.set_ylabel("Production (in tons)")
    ax.set_xlabel("State")
    ax.legend(title="Crops", bbox_to_anchor=(1.05, 1), loc='upper left')

    plt.xticks(rotation=45)
    plt.tight_layout()

    return fig

# Bind function to Panel
interactive_plot = pn.bind(plot_stacked_bar, crop_selector, state_selector)

# Create Panel dashboard layout
dashboard = pn.Row(
    "# Top 10 States & Crops by Production",
    "### Select Crops:",
    state_selector,
    crop_selector,
    interactive_plot
).servable()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import panel as pn
import seaborn as sns

# Initialize Panel
pn.extension()

# Load CSV file
file_path = "Gujarat.csv"  # Update if needed
df2 = pd.read_csv(file_path)

# Ensure 'QueryType' column exists
if "QueryType" not in df2.columns:
    raise ValueError("Column 'QueryType' not found in the dataset!")

# Set Seaborn Style for better aesthetics
sns.set_style("whitegrid")
plt.style.use("ggplot")

# Count occurrences of each QueryType
query_counts = df2["QueryType"].value_counts()

# Get top 10 most frequent QueryTypes
N = 10
top_n_query_types = query_counts.nlargest(N).index.tolist()

# Create a new DataFrame with only the top 10 QueryTypes
top_n_data = df2[df2["QueryType"].isin(top_n_query_types)]

# Group data by QueryType and count occurrences
top_n_data = top_n_data["QueryType"].value_counts().reset_index()
top_n_data.columns = ["QueryType", "Count"]  # Rename columns for clarity

# Panel CheckBoxGroup for interactive selection
checkboxes = pn.widgets.CheckBoxGroup(
    name="Select Query Types",
    options=top_n_query_types,  # Options from top 10
    value=top_n_query_types  # Default selection (all selected)
)

# Function to plot Pie Chart dynamically
def plot_pie_chart(selected_options):
    filtered_data = top_n_data[top_n_data["QueryType"].isin(selected_options)]
    
    # If no selection, return empty figure
    if filtered_data.empty:
        return plt.figure(figsize=(10, 6))

    # Plot Pie Chart
    fig, ax = plt.subplots(figsize=(10, 6))
    colors = sns.color_palette("pastel", len(filtered_data))  # Soft color scheme
    
    wedges, _ = ax.pie(
        filtered_data["Count"],
        colors=colors,
        wedgeprops={"edgecolor": "black", "linewidth": 1, "antialiased": True}
    )
    
    # Fixed position legend (right side)
    ax.legend(
        wedges, filtered_data["QueryType"], 
        title="Query Types", 
        loc="center left", 
        bbox_to_anchor=(1, 0.5),  # Always outside the pie chart
        fontsize=10, 
        frameon=True
    )
    
    plt.title("Query Types in Kisan Call Centre Scheme", fontsize=14, fontweight="bold")
    
    return fig

# Bind function to Panel UI for dynamic updates
pie_chart_pane = pn.bind(plot_pie_chart, checkboxes)

# Create the Panel layout
dashboard = pn.Column(
    pn.pane.Markdown("### Select Query Types to Update Pie Chart"),
    checkboxes,
    pie_chart_pane
)

# Show the dashboard
dashboard.servable()
```
